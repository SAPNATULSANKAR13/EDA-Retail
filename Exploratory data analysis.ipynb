# Data Visualization

```python
plt.figure(figsize=(10,5))
plt.bar('Sub-Category', 'Category', data=data1)
plt.title('Category VS Sub Category')
plt.xlabel('Sub-Category')
plt.ylabel('Category')
plt.xticks(rotation=45)
plt.show()
```

```python
data1.corr()
```

```python
data1.hist(bins=20, figsize=(20,15))
plt.show()
```

From the above graphs, we can say that the our data is not normal as histogram graph.

```python
plt.figure(figsize=(10,8))
data['Region'].value_counts().plot.pie(autopct = '%1.1f%%')
plt.show()
```

form graph, the west region is highest.

**Profit VS Sales**

```python
fig,ax=plt.subplots(figsize=(16,6))
ax.scatter(data['Sales'],data['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

**Discount VS Profit**

```python
sns.scatterplot(x='Discount', y='Profit', label='Profit', data=data)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x=data1['State'])
plt.xticks(rotation=90)
plt.title("STATE")
plt.show()
```

```python
figsize=(15,10)
sns.pairplot(data1,hue='Sub-Category')
plt.show()
```

**Profit VS Quantity**

```python
sns.lineplot(x='Quantity',y='Profit', label='Profit', data=data)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment wise Sales in each Region')
sns.barplot(x='Region', y='Sales', data=data,hue='Segment',order=data['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region', fontsize=15)
plt.show()
```

segment wise sales are almost same in every region

```python
data.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['brown','green'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

Profit and Sales are maximum in west region.

```python
t_states = data['State'].value_counts().nlargest(10)
t_states
```

High profit is for California and New York.

```python
data.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['blue','purple'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```
